```python
indices=[0,200,400,600,800,1000,1200,1400,1600,1800,1999]
directory = "/media/miplab-nas2/Data3/Hamid_Alicia"
modes = 2000
hemispheres = ['rh', 'lh']
mat_dir = '2000_modes' 
name_col = [f"SSBCAP_{index}" for index in range(1,401)]
subjects=subjects_HCP
txt_norma="norma_eval_2000.txt"
inter_=200
main_df_corr, lambda_inter=get_corr_df(subjects, directory,modes,indices, hemispheres, mat_dir, name_col, txt_norma, inter_)

# Only keep the right hemipshere 
SSBCAPS=[f"SSBCAP_{i+1}" for i in range(400)]
main_df_corr=main_df_corr[main_df_corr.Map.isin(SSBCAPS[200:400])]

# get the NW and add ir as a column
nw=pd.read_csv("/media/miplab-nas2/Data3/Hamid_Alicia/nw_atlas/ParcelNetworkName.csv", sep= ",", header=None)
nw_right=nw[nw.columns[200:400]]
nw_right.columns = [f'SSBCAP_{int(col) + 1}' for col in nw_right.columns]
mapping = {col: nw_right[col].iloc[0] for col in nw_right.columns}
main_df_corr['nw'] = main_df_corr['Map'].map(mapping)

#separate left and right hemisphere 
bound = loadmat('/media/miplab-nas2/Data3/Hamid_Alicia/splitting/lmax.mat')['lmax'][0][0] #HCP
left_all=main_df_corr[main_df_corr['Hemisphere']=='left']
right_all=main_df_corr[main_df_corr['Hemisphere']=='right']

# Per Network --> 7 different plots, each representing the mean accuracy for each subject, 
#for the maps associted to one NW in paricular

group_nw_left=left_all.groupby(['subject','eigen_value' , 'nw'])['accuracy'].mean().reset_index()
group_nw_right=right_all.groupby(['subject', 'eigen_value','nw'])['accuracy'].mean().reset_index()
```

### Get the Spatial Map from the Subband - BioFinder

```python
def save_recon_subband(sub_band,  subject, dir_,) : 
    betas_r=mat73.loadmat(f"{dir_}/{subject}/500_modes/rh_betas_all_mode_500_normed.mat")['recon_beta_final']
    modes_r =mat73.loadmat(f"{dir_}/{subject}/500_modes/rh_eigen_wo_nan_500.mat")['eigenmodes']
    indexes_r= loadmat(f"{dir_}/{subject}/500_modes/rh_grouped_values_diff.mat")['Y2'].flatten()
    indexes_list = [list(arr.flatten() - 1) if arr.size > 0 else [] for arr in indexes_r]


    beta_selected_r=betas_r[indexes_list[sub_band]]
    reconstruction_r= modes_r[:,indexes_list[sub_band]].dot(beta_selected_r)
    savemat(f"{dir_}/{subject}//500_modes/rh_{sub_band+1}_recon_recon.mat", 
            {'reconstructions':reconstruction_r})

    betas_l=mat73.loadmat(f"{dir_}/{subject}/500_modes/lh_betas_all_mode_500_normed.mat")['recon_beta_final']
    modes_l =mat73.loadmat(f"{dir_}/{subject}/500_modes/lh_eigen_wo_nan_500.mat")['eigenmodes']
    indexes_l= loadmat(f"{dir_}/{subject}/500_modes/lh_grouped_values_diff.mat")['Y2'].flatten()
    indexes_list = [list(arr.flatten() - 1) if arr.size > 0 else [] for arr in indexes_l]

    beta_selected_l=betas_r[indexes_list[sub_band]]
    reconstruction_l= modes_l[:,indexes_list[sub_band]].dot(beta_selected_l)
    savemat(f"{dir_}/{subject}/500_modes/lh_{sub_band+1}_recon_recon.mat", 
            {'reconstructions':reconstruction_l})
    
```
